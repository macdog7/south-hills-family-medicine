# Recover `src/styles/global.css` from Git history

This notebook checks git history and restores a prior version of `src/styles/global.css` if available. The steps are:
1. Check current git status.
2. Locate commits that modified `global.css`.
3. Inspect a previous version.
4. Restore from a chosen commit.
5. Verify and optionally commit.

```python
import subprocess
from pathlib import Path

repo_path = Path('/workspaces/south-hills-family-medicine')
print('Repo path:', repo_path)

def run_git(args):
    result = subprocess.run(['git', '-C', str(repo_path)] + args, capture_output=True, text=True)
    print('> git', ' '.join(args))
    print('returncode:', result.returncode)
    if result.stdout:
        print(result.stdout)
    if result.stderr:
        print(result.stderr)
    return result

status = run_git(['status', '--short'])
history = run_git(['log', '--oneline', '--', 'src/styles/global.css'])

if history.stdout.strip():
    lines = history.stdout.strip().splitlines()
    for idx, line in enumerate(lines[:5], 1):
        print(f'{idx}. {line}')
    commit = lines[0].split()[0]
    print('
Inspecting file from latest commit affecting global.css:')
    show_result = run_git(['show', f'{commit}:src/styles/global.css'])
    if show_result.returncode == 0:
        # Only print first 200 lines to keep output manageable
        print('--- past global.css content ---')
        print('
'.join(show_result.stdout.splitlines()[:200]))
else:
    print('No history found for src/styles/global.css.')
```
